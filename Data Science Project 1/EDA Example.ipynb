```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #visualisation
import seaborn as sns #visualisation
%matplotlib inline
sns.set_color_codes = True
```

```python
df = pd.read_csv("D:\Data Science\iris data set\Car_data.csv")
df.head()
```

```python
df.tail()
```

```python
df.dtypes
```

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity','Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

```python
df = df.rename(columns={'Engine HP': 'HP', 'Engine Cylinders': 'Cylinders', 'Transmission Type': 'Transmission Type',
                        'Driven_Wheels': 'Drive Mode','highway MPG': 'MPG-H', 'city mpg': 'MPG-C', 'MSRP': 'Price' })
df.head(5)
```

```python
# Total number of rows and column
df.shape

duplicate_df = df[df.duplicated()]

duplicate_df.shape

```

```python
df.count()
# df
```

```python
df= df.drop_duplicates()
df.head()
```

```python
df.count()
```

```python
# df
```

```python
# Dropping the missing or null values.

# Finding the null values.
print(df.isnull().sum())
```

```python
# Dropping the missing values.
df = df.dropna() 
df.count()
# df
```

```python
# After dropping the values
print(df.isnull().sum()) 
```

```python
#Detecting Outliers
#The outlier detection and removing that I am going to perform is called IQR score technique.

sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1
print(IQR)
```

```python
print(type(Q1))
print(type(Q3))
print(type(IQR))


for x in Q1:
    print(type(x))   
for x in Q3:
    print(type(x))    
for x in IQR:
    print(type(x))
```

```python
df.dtypes
```

```python
df.columns
lis_string=[]
lis_numeric=['HP',Cylinders]
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis = 1)]
df.shape

# print((df < (Q1 - 1.5 * IQR))|(df > (Q3 + 1.5 * IQR)))
# df.shape
```

```python
# Plotting a Histogram

df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title('Number of cars by make')
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
# Finding the relations between the variables heat map.

plt.figure(figsize=(20,10))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```
