```python
# Remove punctuation

string.punctuation
```

```python
def remove_punctuation(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation]) # it will discard all the punctuations
    return text_nopunct

data['body_text_clean'] = data['text'].apply(lambda x : remove_punctuation(x))
data.head()
```
